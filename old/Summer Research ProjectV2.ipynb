```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the Excel file using pandas
data = pd.read_excel('reserves_currency_share2021.xlsx')

CountryColumn = data['Country']
CountryColumn
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

def visualize_countries_presence(countries_present):
    # Read the world map shapefile
    world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

    # Create a new column to mark countries present or not
    world['Presence'] = world['name'].apply(lambda country: 'Present' if country in countries_present else 'Not Present')

    # Plot the world map
    fig, ax = plt.subplots(figsize=(15, 10))
    world.plot(column='Presence', cmap='coolwarm', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

    # Set the title
    ax.set_title('Countries in Dataset')

    # Show the plot
    plt.show()

# Example usage
country_list = CountryColumn
visualize_countries_presence(country_list)
```
